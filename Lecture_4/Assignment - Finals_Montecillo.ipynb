```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.tree import minimum_spanning_tree
from networkx.algorithms.shortest_paths.weighted import dijkstra_path

# Full name: Matthieu Blue D. Montecillo
# Subject: CIS1203 Discrete Math 2
# Group #: 1

# Create the graph
G = nx.Graph()
G.add_nodes_from(["A","B","C","D","E","F","G"])
edgelist = [("A","C"),("A","I"),("C","B"),("C","E"),("C","D"),("B","E"),("I","E"),("E","F"),("F","G"),("G","E")]
G.add_edges_from(edgelist)

# Plot the original graph
plt.figure(figsize=(10, 8))
pos = nx.planar_layout(G)
nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={("A","C"):10,("A","I"):5,("C","B"):20,("C","E"):15,("C","D"):30,("B","E"):25,("I","E"):40,("E","F"):35,("F","G"):50,("G","E"):45},
    font_color='red'
)
plt.title('Original Graph')
plt.show()

# Apply Kruskal's Algorithm to find the minimum spanning tree
MST_kruskal = minimum_spanning_tree(G)

# Plot the minimum spanning tree using Kruskal's Algorithm
plt.figure(figsize=(10, 8))
nx.draw_networkx_edges(G, pos, edgelist=MST_kruskal.edges(), edge_color='green')
nx.draw_networkx_nodes(G, pos, node_size=500, node_color='pink', alpha=0.9)
nx.draw_networkx_labels(G, pos)
plt.title("Minimum Spanning Tree (Kruskal's Algorithm)")
plt.show()

# Apply Prim's Algorithm to find the minimum spanning tree
MST_prim = nx.minimum_spanning_tree(G)

# Plot the minimum spanning tree using Prim's Algorithm
plt.figure(figsize=(10, 8))
nx.draw_networkx_edges(G, pos, edgelist=MST_prim.edges(), edge_color='blue')
nx.draw_networkx_nodes(G, pos, node_size=500, node_color='pink', alpha=0.9)
nx.draw_networkx_labels(G, pos)
plt.title("Minimum Spanning Tree (Prim's Algorithm)")
plt.show()

# Find the shortest path from node A to node G using Dijkstra's Algorithm
shortest_path = dijkstra_path(G, source="A", target="G")

# Plot the shortest path
plt.figure(figsize=(10, 8))
nx.draw_networkx_edges(G, pos, edgelist=[(shortest_path[i], shortest_path[i+1]) for i in range(len(shortest_path)-1)], edge_color='purple', width=2)
nx.draw_networkx_nodes(G, pos, node_size=500, node_color='pink', alpha=0.9)
nx.draw_networkx_labels(G, pos)
plt.title("Shortest Path (Dijkstra's Algorithm)")
plt.show()
```
