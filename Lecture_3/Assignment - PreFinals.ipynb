```python
import networkx as nx
import matplotlib.pyplot as plt

# Full name: Matthieu Blue D. Montecillo
# Subject: CIS1203 Discrete Math 2
# Group #: 1

# Create a directed graph
G = nx.DiGraph()
G.add_edges_from([(1, 2), (2, 3), (3, 1)])

# Plot the directed graph
pos = nx.spring_layout(G)  # Compute node positions
nx.draw(G, pos, with_labels=True, arrows=True)
plt.title('Directed Graph')
plt.text(0, -1.5, "This graph represents a directed relationship between nodes. Each edge has a direction indicated by an arrow, showing the flow or direction of the relationship between nodes.", fontsize=10, ha='center')
plt.show()
```

```python
# Create a simple graph
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 1)])

# Plot the simple graph
pos = nx.spring_layout(G)  # Compute node positions
nx.draw(G, pos, with_labels=True)
plt.title('Simple Graph')
plt.text(0, -1.5, "A simple graph is an undirected graph where no loops (edges from a node to itself) or multiple edges between the same pair of nodes exist. It represents a basic network without directed relationships.", fontsize=10, ha='center')
plt.show()
```

```python
# Create a complete graph
G = nx.complete_graph(5)

# Plot the complete graph
pos = nx.circular_layout(G)  # Compute node positions
nx.draw(G, pos, with_labels=True)
plt.title('Complete Graph')
plt.text(0, -1.5, "In a complete graph, every pair of distinct nodes is connected by a unique edge. This means that there are no isolated nodes, and every node is directly connected to every other node in the graph.", fontsize=10, ha='center')
plt.show()
```

```python
# Create a bipartite graph
G = nx.complete_bipartite_graph(2, 3)

# Plot the bipartite graph
pos = nx.spring_layout(G)  # Compute node positions
nx.draw(G, pos, with_labels=True)
plt.title('Bipartite Graph')
plt.text(0, -1.5, "A bipartite graph is a graph whose nodes can be divided into two disjoint sets such that no edge exists between nodes within the same set. In this case, a complete bipartite graph is created with two sets of nodes, where every node in one set is connected to every node in the other set.", fontsize=10, ha='center')
plt.show()
```

```python
# Create a subgraph
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5)])

# Create a subgraph
sub_nodes = [1, 2, 3]
sub_G = G.subgraph(sub_nodes)

# Plot the subgraph
pos = nx.spring_layout(sub_G)  # Compute node positions
nx.draw(sub_G, pos, with_labels=True)
plt.title('Subgraph')
plt.text(0, -1.5, "A subgraph is a subset of the original graph that includes a selection of nodes and the edges connecting only those nodes. In this case, a subgraph of nodes 1, 2, and 3 is created from a larger graph.", fontsize=10, ha='center')
plt.show()
```

```python
# Create a graph with an Eulerian circuit
G = nx.cycle_graph(5)

# Plot the graph with Euler's circuit
pos = nx.circular_layout(G)  # Compute node positions
nx.draw(G, pos, with_labels=True)
plt.title("Euler's Circuit")
plt.text(0, -1.5, "An Eulerian circuit is a circuit that traverses every edge of the graph exactly once and returns to the starting node. The graph is structured in such a way that it forms a closed loop without repeating any edges.", fontsize=10, ha='center')
plt.show()
```

```python
# Create a graph with an Eulerian trail
G = nx.path_graph(5)

# Plot the graph with Euler's trail
pos = nx.circular_layout(G)  # Compute node positions
nx.draw(G, pos, with_labels=True)
plt.title("Euler's Trail")
plt.text(0, -1.5, "An Eulerian trail is a path that traverses every edge of the graph exactly once but may not return to the starting node. Unlike an Eulerian circuit, an Eulerian trail starts and ends at different nodes.", fontsize=10, ha='center')
plt.show()
```

```python
# Create a graph with a Hamiltonian circuit
G = nx.cycle_graph(5)

# Plot the graph with Hamiltonian circuit
pos = nx.circular_layout(G)  # Compute node positions
nx.draw(G, pos, with_labels=True)
plt.title("Hamiltonian Circuit")
plt.text(0, -1.5, "A Hamiltonian circuit is a circuit that visits every node of the graph exactly once and returns to the starting node. Unlike Eulerian circuits, Hamiltonian circuits do not require traversing every edge but instead focus on visiting every node.", fontsize=10, ha='center')
plt.show()
```
